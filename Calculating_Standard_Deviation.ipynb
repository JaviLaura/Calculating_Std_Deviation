```python
def std(df,colName):
  x_bar = df[colName].mean()
  sum = 0
  n = df[colName].count()
  for measurement in df[colName]:
    sum += (measurement - x_bar) ** 2
  
  return math.sqrt(sum/(n-1))


print(std(HWdf,"Height"))
print(std(HWdf,"Weight"))
```

**Part 2: Calculate Percent Values**

Write a function `percentage()` that calculates the percentage of values in a distribution that fall between lower and upper bound.

Then test your function and ensure that roughly 68% falls within 1 standard deviation of the mean, 96% falls within 2 standard deviations, and 99.7 falls within 3 standard deviations. 

```python
def percentage(df,colName,low,high):
  inRange = df[(df[colName]<high) & (df[colName]>low)]
  ratio = (inRange[colName].count())/(df[colName].count())
  percent = ratio * 100
  return str(round(ratio * 100,3)) + "%"

mean = HWdf["Height"].mean()
std = HWdf["Height"].std()

lowBound = mean - (2 * std)
highBound = mean + (2 * std)
#Replace num for as many std you want
print(percentage(HWdf,"Height",lowBound,highBound))
  
```
